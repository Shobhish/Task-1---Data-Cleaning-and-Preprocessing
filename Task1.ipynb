```python
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["director"]=df["director"].fillna("Unknown")
df["cast"]=df["cast"].fillna("Unknown")
df["country"]=df["country"].fillna("Unknown")
df["rating"]=df["rating"].fillna("Not Rated")
```

```python
df=df.dropna(subset=["date_added"])
```

```python
df.duplicated().sum()
```

```python
text_columns = ["director" ,"cast", "country", "type", "rating","date_added"]
for col in text_columns:
    df[col]=df[col].str.strip()
```

```python
df["date_added"]=pd.to_datetime(df["date_added"])
```

```python
df.dtypes
```

```python
df.columns=df.columns.str.lower().str.strip().str.replace(" ","_")
print(df.columns)
```

```python
df.to_csv("cleaned_netflix_titles.csv", index=False)
```

```python
Data Cleaning & Preprocessing Summary
Loaded the Netflix Movies and TV Shows dataset using Pandas.
Identified missing values using isnull().sum().
Handled missing values by filling missing entries in director, cast, country, and rating, and removing rows with missing date_added.
Checked for duplicate records using duplicated().sum(). No duplicate rows were found.
Standardized text columns by removing leading and trailing spaces using str.strip().
Converted the date_added column to datetime format using pd.to_datetime().
Verified that the column headers were already clean and consistent (lowercase with underscores).
Checked the data types using dtypes and confirmed they were correct after converting date_added.
Saved the cleaned dataset as cleaned_netflix_titles.csv.
```
